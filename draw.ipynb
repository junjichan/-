```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 设置中文字体（解决中文显示问题）
plt.rcParams['font.sans-serif'] = ['SimHei']  # Windows系统常用字体
# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # Mac系统常用字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

data1 = pd.read_csv('./1.原始版本/train_process.csv')
data2 = pd.read_csv('./2.损失函数修改/第一种损失函数/train_process.csv')
data3 = pd.read_csv('./2.损失函数修改/第二种损失函数/train_process.csv')
data4 = pd.read_csv('./2.损失函数修改/nce损失函数/train_process.csv')
data5 = pd.read_csv('./3.数据增强版本/train_process.csv')
```

```python
# 创建图像
plt.figure(figsize=(10, 4))

# 绘制训练损失和验证损失曲线
plt.plot(data1['epoch'], data1['train_loss'], label='Train Loss', color='blue')
plt.plot(data1['epoch'], data1['val_loss'], label='Validation Loss', color='red')

# 添加标题和标签
plt.title('Cross Entropy损失曲线')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.ylim(0, 1)

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```
